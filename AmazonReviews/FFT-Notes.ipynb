```python
# Imports
import numpy as np
import matplotlib.pyplot as plt 


# Sine Wave Function
def sine_wave(f, a, duration, sample_rate): #f=frequency & a=the range of potential y-values
    t = np.arange(0, duration, 1/sample_rate)
    return a * np.sin(2 * np.pi * f * t), t

# Call function
s, t = sine_wave(f=261.63, 
                   a=1,
                   duration=0.05,
                   sample_rate=10e3)
# returns sine_wave(t) & t = (s, t)

# Plot Waves
fig = plt.figure(figsize=(28, 2.5))
#plt.plot(t, s, color="#95B0FF")
plt.scatter(t, s, color="#95B0FF")
```
